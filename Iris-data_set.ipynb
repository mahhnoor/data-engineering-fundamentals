```python
import pandas as pd

Data1= {
    'species': ['setosa', 'versicolor', 'virginica', 'setosa'],
    'sepal_length': [5.1, 6.3, 7.2, 4.9]
}
df_measure= pd.DataFrame(Data1)

Data2= {
    'species': ['setosa', 'versicolor', 'virginica'],
    'code': ['SET', 'VER', 'VIR']
}
df_codes= pd.DataFrame(Data2)

print("----Data Frame 1:----")
print(df_measure)

print("\n----Data Frame 2:----")
print(df_codes)
```

```python
merged_df= pd.merge(df_measure, df_codes, on= 'species', how= 'inner')

print("---- Merged Data Frame ----")
print(merged_df)
```

```python
left_join_df= pd.merge(df_measure, df_codes, on= 'species', how= 'left')

print("---- Left Joined Data Frame ----")
print(left_join_df)
```

```python
import sqlite3
```

```python
conn= sqlite3.connect('practice_db.sqlite')
print("Database Connection Established")
```

```python
#Write the DataFrame 'df' to a SQL table named 'iris_data'
# index=False means don't save the Pandas row index
# if_exists='replace' means if the table already exists, overwrite it

df.to_sql('iris_data', conn, if_exists= 'replace', index='False')
print("Data Frame loaded into 'iris_data' table.")
```

```python
sql_query= """
SELECT species, AVG(sepal_length) AS avg_sepal_length
FROM 'iris_data'
GROUP BY species;
"""

results_df= pd.read_sql(sql_query, conn)

__builtins__.print("Results Extracted from SQL")
__builtins__.print(results_df)
```

```python
conn.close()
__builtins__.print("Database Connection Closed")
```
